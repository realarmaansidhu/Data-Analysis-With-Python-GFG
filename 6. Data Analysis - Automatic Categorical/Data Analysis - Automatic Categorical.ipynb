# Data Analysis - Automatic Categorical

```python
categories, types, content_rating = {}, {}, {}

for name in df["Category"].unique():
    counter = 0
    for cats in df["Category"]:
        if(cats == name):
            counter += 1
    categories[name] = counter

print(categories, end="\n")

for type in df["Type"].unique():
    counter = 0
    for cats in df["Type"]:
        if(cats == type):
            counter += 1
    types[type] = counter

print(types, end="\n")

for cr in df["Content Rating"].unique():
    counter = 0
    for cats in df["Content Rating"]:
        if(cats == cr):
            counter += 1
    content_rating[cr] = counter

print(content_rating)
```

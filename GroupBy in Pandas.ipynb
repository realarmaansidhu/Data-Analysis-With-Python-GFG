```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv("googleplaystore.csv")

imputer = SimpleImputer(missing_values = np.NaN, strategy = "mean")
df.iloc[:, 2:3] = np.round(imputer.fit_transform(df.iloc[:, 2:3]), 2)
df.info()
```

## Q1 Name the apps in ART_AND_DESIGN with rating more than 4.5 in ascending order

```python
df[(df["Category"] == "ART_AND_DESIGN") & (df["Rating"] >= 4.50)].sort_values(by=["Rating"])
```

## Q2 Name the apps in ART_AND_DESIGN in descending order WRT Ratings

```python
df[df["Category"] == "ART_AND_DESIGN"].sort_values(by=["Rating"], ascending = False)
```

## Q3 Which Category is having Maximum Average Rating?

```python
df.groupby(by=["Category"])["Rating"].mean().idxmax()
```

## Q4 How many Free Apps are in each Category in Descending Order?

```python
df[df["Type"] == "Free"].groupby(by=["Category"]).count()["Type"].sort_values(ascending=False)
```
